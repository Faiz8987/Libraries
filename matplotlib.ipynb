```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(2,22,2)
```

```python
x
```

```python
a=x*2
```

```python
b=x*3
```

```python
plt.plot(x,a,linestyle=":",color="g")
plt.plot(x,b,linewidth="5",color="r")
plt.title("kolo")
plt.xlabel("jojo")
plt.ylabel('lolo')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,a,linewidth=8,color="b")
plt.title("ggg")
plt.xlabel("jjkjk")
plt.ylabel("jdhkjk")
plt.grid(True)


plt.subplot(1,2,2)
plt.plot(x,b,linestyle=":",color="g")
plt.title("ggg")
plt.xlabel("jjkjk")
plt.ylabel("jdhkjk")
plt.grid(True)
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[9,8,7,6,5,4,3,2,1]
f=[2,5,8,9,6,3,1,4,7]
```

```python
plt.scatter(x,y,marker="*",color="r",s=80)
plt.scatter(x,f,marker=".",color="g",s=60)
plt.title("jojo")
plt.xlabel("hhh")
plt.ylabel("ddd")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,color="b",marker=".",s=70)
plt.title("tit")
plt.xlabel("joo")
plt.ylabel("ggg")

plt.subplot(1,2,2)
plt.scatter(x,f,color="g",marker="*",s=90)
plt.title("tit")
plt.xlabel("joo")
plt.ylabel("ggg")
plt.show()
```
